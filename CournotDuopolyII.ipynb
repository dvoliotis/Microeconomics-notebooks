##  Δυοπώλιο  Cournot  με ασύμμετρο κόστος

```python
from sympy import *
from sympy import init_printing # Το init_printing εισάγει το LaTeX printing
init_printing(use_latex='mathjax')
import math
import matplotlib.pyplot as plt
```

Παράμετροι υποδείγματος

```python
# Παράμετροι για την αντίστροφη συνάρτηση ζήτησης
A, b = var('A, b', real=True, positive=True)
# Ποσότητες
Q, q1 , q2 = var('Q, q1, q2', real=True, positive=True)
# Κόστος, τιμή
T, p = var('T, p', real=True, positive=True)
```

Συνάρτηση κερδών

```python
# Ορισμός αντίστροφης συνάρτηση ζήτησης
inverse = p- A + b*Q
inv =inverse.xreplace({Q :q1+q2})
price = solve(inv, p)



price
```

Συνάρτηση κόστους για τις δύο επιχειρήσεις

```python
totalcost1 = 4*q1
totalcost2 = 2*q2 + T # το Τ εκφράζει το κόστος απόκτησης τεχνολογίας που απαιτείται ώστε να παράγει με χαμηλότερο κόστος
```

Συνάρτηση κερδών γα τις δύο επιχειρήσεις

```python
profit1 = price[0]*q1 - totalcost1
profit2 = price[0]*q2 - totalcost2

profit1, profit2
```

Συναρτήσεις βέλτιστης αντίδρασης

```python
#Συνάρτηση βέλτιστης αντίδρασης της Επιχείρησης 1
Bresponse1 = diff(profit1, q1)
# Συνάρτηση βέλτιστης αντίδρασης της Επιχείρησης 2
Bresponse2 =diff(profit2, q2)

BR1 =solve(Bresponse1, q1)
BR2 = solve(Bresponse2, q2)

BR1, BR2
```

Ισορροπία  Nash


```python
Nash = solve((q1-BR1[0], q2-BR2[0]), q1, q2)
Nash
```

Μια αριθμητική λύση

 Ισορροπία Nash


```python
vals = {A:100, T:10, b:1}
nNash = list(Nash.values())

nNash[0].subs(vals), nNash[1].subs(vals)
```
